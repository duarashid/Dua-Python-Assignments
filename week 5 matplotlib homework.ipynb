```python
import matplotlib.pyplot as plt

fruits = ['avocados', 'pineapples','apples','peaches','bananas','tomatoes','oranges']
values = [2, 1, 3, 2, 6, 4, 5]
color = ['pink']
plt.fill_between(fruits, values, color=color)
plt.title('Fruits in my fridge')
plt.xlabel('fruits')
plt.ylabel('values')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['avocados', 'pineapples','apples','peaches','bananas','tomatoes','oranges']
sizes = [2, 1, 3, 2, 6, 4, 5]
colors = ['#919866','#FDD773','#C7372F','#FFE5B4','#ffe135','#FF6347','#FFA500']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Fruits in my fridge')
plt.show()
```
